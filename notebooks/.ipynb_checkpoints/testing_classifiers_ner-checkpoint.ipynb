```python
import pandas as pd
from transformers import pipeline
from transformers import AutoTokenizer, AutoModelForTokenClassification, pipeline
import string
```

```python
#NER

ner_model = AutoModelForTokenClassification.from_pretrained('carolanderson/roberta-base-food-ner')
tokenizer = AutoTokenizer.from_pretrained("roberta-base", add_prefix_space=True)

nlp = pipeline("ner", model=ner_model, tokenizer=tokenizer)
```

```python
# Text classifier mode

classifier = pipeline("zero-shot-classification",
                      model="MoritzLaurer/deberta-v3-base-zeroshot-v1.1-all-33")
```

## Workflow


1. Get a test_str input 
2. pass it into NER for FOOD 
3. make those food words go into classification.
4. the minute you encounter a non-veg item --> break and it's non veg 

### 1. Get a test_str input 

```python
test_str = input("What menu items do you want to classify?")
test_str
```

### 2. pass it into NER for FOOD

```python
#list of dictionaries
menu_ner = nlp(test_str, aggregation_strategy="first")
menu_ner
```

```python
# put them into a list of food NERs

food_ners = []

for item in menu_ner:
    
    food_ners.append(item["word"].strip(string.punctuation))
    
food_ners
```

### 3. make those food words go into classification.

```python
results = classifier(food_ners, ['vegetarian', 'meat', 'seafood'])
labels = [result["labels"][0] for result in results] # should be a list
```

```python
labels
```

### 4. the minute you encounter a non-veg item --> break and it's non veg 

```python
if 'meat' in labels or 'seafood' in labels:
    print("Not vegetarian")
else:
    print("vegetarian")
```

## Put it into a nice function to do all of this

- take input a string for now and then you can change for csv

```python
def get_diet():
    
    """
    
    Asks for a string input from user and then returns whether or not it's vegetarian
    
    
    
    """
    
    #NER
    ner_model = AutoModelForTokenClassification.from_pretrained('carolanderson/roberta-base-food-ner')
    tokenizer = AutoTokenizer.from_pretrained("roberta-base", add_prefix_space=True)

    nlp = pipeline("ner", model=ner_model, tokenizer=tokenizer)

    # Text classifier mode
    classifier = pipeline("zero-shot-classification",
                          model="MoritzLaurer/deberta-v3-base-zeroshot-v1.1-all-33")
    
    #get input string
    test_str = input("What menu items do you want to classify?")

    #list of dictionaries
    menu_ner = nlp(test_str, aggregation_strategy="first")

    # put them into a list of food NERs
    food_ners = []
    for item in menu_ner:
        food_ners.append(item["word"].strip(string.punctuation))
        
    #classify them
    results = classifier(food_ners, ['vegetarian', 'meat', 'seafood'])
    labels = [result["labels"][0] for result in results] 

    if ['meat' in labels:
        print("Not Vegetarian")
    else:
        print("Vegetarian")
```

```python
df = pd.read_csv("Menu_items.csv")
df.head()
```

```python
def get_diet_new(test_str, model):
    
    """
    
    Need to pass a test string and the model we want to use to classify as veg or not veg
    
    
    
    """
    
    #NER
    ner_model = AutoModelForTokenClassification.from_pretrained('carolanderson/roberta-base-food-ner')
    tokenizer = AutoTokenizer.from_pretrained("roberta-base", add_prefix_space=True)

    nlp = pipeline("ner", model=ner_model, tokenizer=tokenizer)

    # Text classifier mode
    classifier = pipeline("zero-shot-classification",
                          model=model)
    
    #get input string
    #test_str = input("What menu items do you want to classify?")

    #list of dictionaries
    menu_ner = nlp(test_str, aggregation_strategy="first")

    # put them into a list of food NERs
    food_ners = []
    for item in menu_ner:
        food_ners.append(item["word"].strip(string.punctuation))
        
    #classify them
    results = classifier(food_ners, ['vegetarian', 'not vegetarian'])
    labels = [result["labels"][0] for result in results] 

    if 'meat' in labels:
        return "Not Vegetarian"
    else:
        return "Vegetarian"
```

```python
def classify_menu_items(menus_df, model):
    
    
    texts = menus_df["test_str"].to_list()
    results = [get_diet_new(text, model) for text in texts]
    print("done", model)
    
    return results
        
```

```python
df["bart_classification"] = classify_menu_items(df, "facebook/bart-large-mnli")
df["mDeBERTa-v3-base-mnli-xnli"] = classify_menu_items(df, "MoritzLaurer/mDeBERTa-v3-base-mnli-xnli")

df["DeBERTa-v3-base-mnli-fever-anli	"] = classify_menu_items(df, "MoritzLaurer/DeBERTa-v3-base-mnli-fever-anli")
df["deberta-v3-base-zeroshot-v1.1-all-33"] =classify_menu_items(df, "MoritzLaurer/deberta-v3-base-zeroshot-v1.1-all-33")
```

```python
column_names = df.columns.to_list()[5:]
column_names
```

```python
def get_model_success(column_name):
    
    len_success = df[df["Manual"] == df[column_name]].shape[0]
    len_all = df.shape[0]
    
    success_pc = (len_success * 100) / len_all
    
    return success_pc
```

```python
success_rate = {}

for name in column_names:
    success_rate[name] = get_model_success(name)

    
    
model_success_df = pd.DataFrame()
model_success_df["models"] = list(success_rate.keys())
model_success_df["success_rate"] = list(success_rate.values())
model_success_df.sort_values(by = "success_rate", ascending=False)
```

```python
model_success_df.to_csv("success_rate_ner_vegnotveg.csv", index = False)
```
